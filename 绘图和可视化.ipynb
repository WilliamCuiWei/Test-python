```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
```

### Figure和Subplot

```python
fig = plt.figure()  #创建一个新的figure
ax1 = fig.add_subplot(2,2,1)  #用add_subplot 创建一个或多个subplot,图像是2*2的，当前是4个中的第一个
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

from numpy.random import randn
plt.plot(randn(50).cumsum(),'k--')  #在最后一个图绘图

_=ax1.hist(randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30) + 3*randn(30))
ax4.plot(randn(50),'k')
```

```python
fig,axes = plt.subplots(2,3)
ax1 = fig.add_subplot(2,3,2)
from numpy.random import randn
ax1.plot(randn(50).cumsum(),'k--') 
```

#### 调整subplot周围的间距

```python
fig,axes = plt.subplots(2,2,sharex = True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
```

### 设置标题、轴标签、刻度、图例、刻度标签

```python
fig = plt.figure();ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
```

```python
fig = plt.figure();ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum())
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(['one','two','three','four','five'],rotation = 30,fontsize = 'small')
ax.set_title("My first matplotlib plot")
ax.set_xlabel('Stages')
ax.set_ylabel('Numbers')
```

添加图例

```python
fig = plt.figure();ax = fig.add_subplot(1,1,1)
ax.plot(randn(500).cumsum(),'k',label='one')
ax.plot(randn(500).cumsum(),'k--',label='two')
ax.plot(randn(500).cumsum(),'k.',label='three')
ax.legend(loc = 'best') #ax.legend() 或者plt.legend()自动创建图例
ax.text(100,20,'Hello world!',family='monospace',fontsize = 10)  #在图上注解
```

在图表中添加一个图形

```python
fig = plt.figure();ax = fig.add_subplot(1,1,1)
rect = plt.Rectangle((0.2,0.75),0.4,0.15, color = 'r',alpha = 0.3)
circ = plt.Circle((0.7,0.2),0.15, color = 'b',alpha = 0.3)
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]], color = 'g',alpha =0.1)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
plt.savefig('C:\Users\lenovo\Desktop\ss.png')  #将图表保存到文件
```

### pandas 中的绘图函数

#### 线型图

```python
import pandas as pd
from pandas import Series,DataFrame
```

```python
s = Series(np.random.randn(10).cumsum(),index = np.arange(0,100,10))
plt.plot(s)
```

```python
s.plot(xlim = [0,100],grid = True)  #使用pandas作图
```

```python
df = DataFrame(np.random.randn(10,4).cumsum(0),columns = ['A','B','C','D'],index = np.arange(0,100,10))
df.plot(style = 'k--',alpha = 0.7, kind = 'line',title = "this is a new picture")  #style线型，alpha填充不透明度，kind图表类型‘line’,‘bar’,'barh','kde'
```

```python
fig = plt.figure();ax = fig.add_subplot(1,1,1)
ax.plot(df)
plt.legend(loc = 'best')
```

#### 柱状图

```python
fig,axes = plt.subplots(2,1)
data = Series(np.random.rand(6),index = list('asvdac'))
data.plot( kind = 'bar',ax=axes[0],color = 'g',alpha = 0.4)
data.plot( kind = 'barh',ax=axes[1],color = 'black',alpha = 0.4)
```

```python
df = DataFrame(np.random.rand(6,4),index = ["one",'two','three','four','five','six'],
              columns = pd.Index(['A','B','C','D'],name = 'Genus'))
df
```

```python
df.plot(kind = 'bar') #"Genus"被用作图例的标题
```

```python
df.plot(kind = 'barh',stacked = True, alpha = 0.7) #stacked = True堆积柱状图
```

```python
data.value_counts().plot(kind = ' bar') #显示series中各值出现的频率
```

#### 直方图和密度图

直方图是一种可以对值频率进行离散化显示的柱状图

```python
df['D'].hist(bins = 10,grid = False)  #bins调节宽度  grid设置是否有网格线
```

```python
df['D'].plot(kind = 'kde',grid = False)  #生成一张密度图（标准混合正态分布KDE）
```

```python
comp1 = np.random.normal(0,1,size = 200)  #N(0,1)
comp2 = np.random.normal(10,2,size = 200)  #N(10,4)
values = Series(np.concatenate([comp1,comp2]))
values.hist(bins = 100,alpha = 0.3,color = 'b',normed = True)
values.plot(kind = 'kde',style = 'k--',grid = True)
```
